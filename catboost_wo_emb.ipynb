```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('prepared_data.csv')
test = pd.read_csv('prepared_test.csv')
```

```python
# transform sale_end_date to day count from min of sale_end_date
data['sale_end_date'] = pd.to_datetime(data['sale_end_date'])
test['sale_end_date'] = pd.to_datetime(test['sale_end_date'])
test['sale_end_date'] = (test['sale_end_date'] - data['sale_end_date'].min()).dt.days
data['sale_end_date'] = (data['sale_end_date'] - data['sale_end_date'].min()).dt.days
```

```python
X_train, y_train = data.drop(['id', 'price', 'actual_price', 'start_date', 'close_date', 'latitude', 'longitude', 'description'], axis=1), data['actual_price']
```

```python
cat_features = ['brand', 'model', 'generation', 'modification', 'equipment', 'body_type', 'color', 'owners_count']
```

```python
from catboost import CatBoostRegressor

model = CatBoostRegressor(loss_function='RMSE', random_seed=42, eval_metric='MAPE', task_type='GPU', devices='0:1', cat_features=cat_features, l2_leaf_reg=1, max_depth=10, iterations=1000, learning_rate=0.05)
```

```python
model.fit(X_train, y_train, verbose=True)
```

```python
pd.Series(model.predict(test.drop(['id', 'description', 'latitude', 'longitude'], axis=1))).to_csv('solution6.csv', header=False, index=False)
```
